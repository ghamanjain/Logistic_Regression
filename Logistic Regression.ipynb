# Feature Scaling

```python
X=df.iloc[:,:8]
Y=df.iloc[:,8:9]
```

# Splitting the Dataset

```python
split_value=round(len(df)-(len(df)*0.25)) #splitting the dataset into two parts
```

```python
x_train=np.array(X[:split_value]) #train part consist of 75% of data
y_train=np.array(Y[:split_value])
x_test=np.array(X[split_value:]) #test part consist of 25% of data
y_test=np.array(Y[split_value:])
print(x_train.shape)
print(y_train.shape[1])
```

# Model for Logistic Regression

```python
class Logistic_Regression:
    def __init__(self,x_train):
        d=x_train.shape[1]
        self.slope=np.zeros((d,1))
        self.intercept=0
    def hypothesis(self,x_train):
        Z=np.dot(x_train,self.slope)+self.intercept  #x_train:m x n   slope:n x 1 
        return Z   # Z:m x 1
    def sigmoid_function(self,Z):
        h=1/(1+np.exp((-Z))) # as hypothesis for logistic h=1/(1+exp(-z))
        return h  #h: m x 1
    def cost(self,h,y_train):
        m=y_train.shape[0] #no.of training examples(row)
        J=(-1/m)*(np.dot(y_train.transpose(),np.log(h))+np.dot((1-y_train).transpose(),np.log(1-h)))  # y_train.tran: 1 x m , h:m x 1
        return J
    def gradient_descent(self,h,x_train,y_train):
        m=y_train.shape[0]
        dm=(1/m)*np.dot(x_train.transpose(),(h-y_train))  # x_train.trans: n x m , h: m x 1 , y_train : m x 1 , dm: n x 1
        dc=(1/m)*np.sum(h-y_train) # h: m x 1 y_train : m x 1
        return dm,dc
    def training_model(self,x_train,y_train,iterations,learning_rate):
        cost_list=[]
        iteration_list=[]
        for i in range(iterations):
            Z=self.hypothesis(x_train)
            h=self.sigmoid_function(Z)
            J=self.cost(h,y_train)
            grad=self.gradient_descent(h,x_train,y_train)
            m_up=grad[0]
            c_up=grad[1]
            self.slope=self.slope-(learning_rate*m_up)
            self.intercept=self.intercept-(learning_rate*c_up)
            if(i%10==0):
                cost_list.append(J[0][0])
                iteration_list.append(i)
        plt.plot(iteration_list,cost_list)
        plt.xlabel('Iterations')
        plt.ylabel('Cost')
        plt.show()
    def predict(self,x_test):
        y_pred=np.dot(x_test,self.slope)+self.intercept
        y_pred=self.sigmoid_function(y_pred)
        x=[]
        for i in y_pred:
            if(i>0.5):
                x.append([1])
            else:
                x.append([0])
        return x
```

```python
logg=Logistic_Regression(x_train)
iterations=2500
learning_rate=0.4
logg.training_model(x_train,y_train,iterations,learning_rate)
y_predd=logg.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
yy = confusion_matrix(y_test,y_predd)
print(yy)
```
